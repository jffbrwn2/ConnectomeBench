```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("jeffbbrown2/ConnectomeBench", "MICrONS, Segment Classification", split="train")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

image_index = 5
front_img = ds[image_index]['option_1_front_image']
side_img = ds[image_index]['option_1_side_image']
top_img = ds[image_index]['option_1_top_image']

plt.figure(figsize=(12, 4))
for i, (img, title) in enumerate(zip([front_img, side_img, top_img], ['Front', 'Side', 'Top'])):
    plt.subplot(1, 3, i + 1)
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
plt.tight_layout()
plt.suptitle(ds[image_index]['class_description'], y = 1.10)
plt.show()
```
